```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
%pip install scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("/Users/junyeow/Desktop/annual_income-sg.csv")
data
```

```python
data.drop([48,49,50], axis=0, inplace=True)
data
```

```python
data.info()
# to see if everything we have is fine
```

```python
def remove_comma(num):
  new_num = num.replace(',','')
  return new_num

data['Singapore'] = data['Singapore'].apply(remove_comma)
data[['Year ','Singapore']] = data[['Year ','Singapore']].apply(pd.to_numeric)
data.info()
```

```python
sns.regplot(x='Year ', y='Singapore', data=data)
plt.show()
# There were some dips in a few areas but still on the rise
```

```python
# regressioan and assign a variable
lri = LinearRegression()
```

```python
# train model 
lri.fit(data[['Year ']].values, data[['Singapore']])
# double brackets is to keep a dataframe
```

```python
lri.predict([[2021]])
```
